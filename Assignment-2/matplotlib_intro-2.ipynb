```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
data = np.linspace(1, 15, 10)
data
print(data)

plt.plot(data)
'''If you provide a single list or array to plot, matplotlib assumes it is a sequence of y values, 
and automatically generates the x values for you. Since python ranges start with 0, 
the default x vector has the same length as y but starts with 0.
Btw, it is a line-graph.'''
```

```python
# Creating an array with 100 values from 0 to 2pi
x = np.linspace(0, 2*np.pi, 100)
# An array with the sine of all the values in array x
y = np.sin(x)
print(x)
print(y)
```

```python
# Produces the figure
#plt.figure()
# Plots the sine function
plt.plot(x,y)
```

```python
# Adds a grid
plt.grid(True)
# Plots the sine function
plt.scatter(x,y)
```

```python
# we may change the color of the plot
# Adds a grid
plt.grid(True)
# Plots the sine function with red dots
plt.plot(x,y,'g*') #- means line, or dash, r means red.
#plt.plot(x,y, 'b^')
```

```python
#we can even change the color of the dots. 

# Adds a grid
plt.grid(True)
# Plots the sine function with red dots
plt.plot(x,y,'r.')
```

```python
plt.grid(True)
# Plots the sine function with red dots
plt.plot(x,y,'y')
```

```python
# we may change the color of the plot
# Adds a grid
plt.grid(True)
# Plots the sine function with red dots
color_list = ['r']*len(x)
color_list.pop(-1)
color_list.append('b')
plt.scatter(x,y, c=color_list) # c for color
```

```python
# Adds a grid
plt.grid(True)
# Plots the sine function with red dots
plt.plot(x,y,'b.')
```

```python
#add legends or caption to a plot.
# Adds a grid. 
# 'pad' is the distance between the words and the graph
plt.grid(True)
# Adds a title
plt.title("Plot of a sine function",
         fontsize = 20,
         pad = 15)
# Adds label for the x-axis
plt.xlabel("x",
          fontsize = 15,
          labelpad = 15)
# Adds label for the y-axis
plt.ylabel("sin(x)",
          fontsize = 15,
          labelpad = 15)
# Plots the sine function with red + signs
plt.plot(x,y,'r+')

```

```python
#let's get stylish. For more info, check this out
#https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
# Changing style to fivethityeight
#plt.style.use('fivethirtyeight')
plt.style.use('ggplot')
# Produces the figure
plt.figure()
# Adds a grid
plt.grid(True)
# Adds a title
plt.title("Plot of a sine function",
         fontsize = 20,
         pad = 15)
# Adds label for the x-axis
plt.xlabel("x",
          fontsize = 15,
          labelpad = 20)
# Adds label for the y-axis
plt.ylabel("sin(x)",
          fontsize = 15,
          labelpad = 15)
# Plots the sine function with red + signs
plt.plot(x,y, "r+")
```

```python
#superposition of two graphs
# Generates an array of cos(x) values
y1 = np.cos(x)
# Produces the figure
plt.figure()
# Adds a grid
plt.grid(True)
# Adds a title
plt.title("Plot of a sine and cosine function")
# Adds label for the x-axis
plt.xlabel("x")
# Adds label for the y-axis
plt.ylabel("y")
# Plots the sine function with red + signs
plt.plot(x,y,'r+')
# Plots the cos function with green crosses
plt.plot(x,y1,'gx')
```

```python
#Chances are that you already know what a cosine and a sine function looks like respectively
#But, what if you don't?
#you need to use 'legends' to be more specific with each graph/function here.
# Produces the figure
plt.figure()
# Adds a grid
plt.grid(True)
# Adds a title
plt.title("Plot of a sine and cosine function")
# Adds label for the x-axis
# Adds label for the y-axis
plt.ylabel("y")
plt.xlabel("x")

# Plots the sine function with red + signs and defines a legend label
plt.plot(x,y,'r+',label="sin(x)")
# Plots the cos function with green crosses and defines a legend label
plt.plot(x,y1,'gx',label="cos(x)")
# Adds a legend in the best position
plt.legend(loc='best')
#plt.legend()
```

```python
# how to change the 'focus' of a plot? Want to zoom in or zoom out? Use plt.xlim().
# Produces the figure
plt.figure()
# Adds a grid
plt.grid(True)
# Adds a title
plt.title("Plot of a sine and cosine function")
# Adds label for the x-axis
plt.xlabel("x")
# Adds label for the y-axis
plt.ylabel("y")
# Defines a region along the x-axis to be displayed
plt.xlim(0, 0.5*np.pi)
#can you try to do the same thing with y-axis?
# Plots the sine function with red + signs and defines a legend label
plt.plot(x,y,'r+',label="sin(x)")
# Plots the cos function with green crosses and defines a legend label
plt.plot(x,y1,'gx',label="cos(x)")
# Adds a legend in the best position
plt.legend(loc='best')
```
